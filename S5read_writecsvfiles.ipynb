# List of Methods to Read a CSV File in Python
    Read CSV file using csv.reader
    Read CSV file using .readlines() function

    Read CSV file using csv.DictReader

# 1.using csv.reader :
        allows you to read data from a CSV file using indexes and 
        ideal for simple CSV files
        treats each line as a list of values separated by the specified delimiter (usually a comma). 
        then iterate through the rows and access individual values using their indices
        Suitable for basic scenarios where you need minimal processing.

```python
import csv
```

```python
with open("iris.csv", "r") as csvfile:
    reader_variable = csv.reader(csvfile, delimiter=",")
    for row in reader_variable:
        print(row)
```

```python
#With open takes 2 arguments file locations and name, r os for read, w is for write.
```

```python
  # delimiter=”,” -> which means it needs to also know how to separate the values
```

# 2.Using readlines():
                The readlines() method in Python reads all lines from a text file and returns them as a list,
                each line is an item in the list.
                less efficient for large files because it loads the entire file into memory.


```python
with open("./iris.csv", "r") as csvfile:
    first_line = csvfile.readline()
    print(first_line)
```

```python
#READLINES() - To read multiple lines.
```

```python
with open("./iris.csv", "r") as csvfile:
    all_lines = csvfile.readlines()
    print(all_lines)
```

# 4.Read CSV file in python using csv.DictReader:
                1.each row is represented as a dictionary with column names as keys.
                2.Useful when you want to access columns by name.

```python
with open('iris.csv', 'r') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
       print(row)
```

# List of Methods to Write a CSV file in python
    1. Write CSV file using csv.writer
    2. Write CSV file using writelines() function
    3. Write CSV file using Pandas
    4. Write CSV file using csv.DictWriter

```python
#1.Using csv.writer
import csv
```

```python
header = ['Name', 'M1 Score', 'M2 Score']
data = [['Alex', 62, 80], ['Bard', 45, 56], ['Joe', 85, 98]]
```

```python
filename = 'Students_Data.csv'
with open(filename, 'w', newline="") as file:
    csvwriter = csv.writer(file) # 2. create a csvwriter object
    csvwriter.writerow(header) # 4. write the header
    csvwriter.writerows(data) # 5. write the rest of the data
```

```python
#Trying to read the file that is written
with open("Students_Data.csv", "r") as csvfile:
    reader_variable = csv.reader(csvfile, delimiter=",")
    for row in reader_variable:
        print(row)
```

# Using .writelines()
            writelines() iterates through each list, 
            converts the list elements to a string, 
            and then writes it to the csv file.

```python
header = 'Name,M1 Score,M2 Score'
data = [['Alex', 62, 80], ['Brad', 45, 56], ['Joey', 85, 98]]
filename = 'Student_scores.csv'

with open(filename, 'w') as file:
    file.write(header + '\n')  # Write the header
    for row in data:
        file.write(','.join(map(str, row)) + '\n')  
```

# using csv.DictWriter

```python
import csv
```

```python

with open('Students_Data.csv', 'w', newline='') as csvfile:
    data = [
        {'Name': 'Alex', 'M1 Score': 62, 'M2 Score': 80},
        {'Name': 'Brad', 'M1 Score': 45, 'M2 Score': 56},
        {'Name': 'Joey', 'M1 Score': 85, 'M2 Score': 98}
    ]
    fieldnames = ['Name', 'M1 Score', 'M2 Score']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(data)
```

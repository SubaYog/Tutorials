# isocalendar() :
    function in Python
    ISO Year: The year according to the ISO standard 8601 and ISO standard 2015. 
    ISO years always start with Monday, and Thursday is considered the middle day of the week.
    ISO Week Number: The week number within the ISO year. 
    ISO weeks can have either 52 full weeks or 53 full weeks.
    ISO Weekday: The weekday corresponding to the date. 
    In the ISO system, Monday is represented as 1, Tuesday as 2, and so on, with Sunday as 7.

ISO calendar system provides a standardized way to represent time

```python
df1['Week'] = df1.Date.dt.isocalendar().week 
df1['Week'].unique()
```

```python
# To get the sales of particular dates eg:2010-05-02, 2010-06-02
```

```python
start_date = '2010-05-02'
end_date = '2010-06-02'
filtered_sales_df = df1[(df1['Date'] >= start_date) & (df1['Date'] <= end_date)]
filtered_sales_df
```

```python
filtered_sales_df.nunique()
```

```python
weekly_sales = filtered_sales_df.groupby(['Store', pd.Grouper(key='Date', freq='W')])['Weekly_Sales'].sum()

# Print the weekly sales for each store
for store, weekly_sales_data in weekly_sales.groupby('Store'):
    print(f"Store {store}:")
    print(weekly_sales_data)
    print("\n")
```

```python
weekly_sales = df1.groupby(['Store', pd.Grouper(key='Date', freq='Y')])['Weekly_Sales'].sum()

# Print the weekly sales for each store
for store, weekly_sales_data in weekly_sales.groupby('Store'):
    print(f"Store {store}:")
    print(weekly_sales_data)
    print("\n")
```
